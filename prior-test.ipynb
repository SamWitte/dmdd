```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import dmdd
xe = dmdd.Experiment('Xe','xenon',5,40,2000,dmdd.eff.efficiency_Xe,0.,1.,energy_resolution=True)
si = dmdd.UV_Model('SI',['mass','sigma_si'],fixed_params={'fnfp_si':1.},GF=False)

```

```python
npoints = 50

ms = np.logspace(0,3, npoints)
ss = np.logspace(-3,5,npoints)


ppointsx,ppointsy = np.meshgrid(ms,ss)
plt.figure()
for i,m in enumerate(ms):
    plt.scatter(m*np.ones(npoints),ss)
    
plt.xlim(xmin=0)
plt.ylim(ymin=0)
```

```python

npointsm = 3
npointss = 5

ms = np.logspace(0,3, npointsm)
ss = np.logspace(0,3,npointss)

ms = [10,20,30,50,100]
ss = [10,100,200,250,300,400,500,600,1000]

for m in ms:
    for s in ss:
        dmdd.Plot_Modulation(xe,[si,si],
                             [{'mass':20.,'sigma_si':100.},{'mass':m,'sigma_si':s}],
                             GF=False,label_params=True)
```
